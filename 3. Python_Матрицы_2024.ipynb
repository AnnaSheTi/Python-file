**Задача 1.** Написать программу решения системы линейных алгебраический уравнений на языке Python, используя модуль numpy. Для тестирования программы решения систем линейных алгебраических уравнений будем использовать следующую тестовую систему:
$$
A \cdot x =b
$$
Матрица $A(N,N)$ формируется следующим образом. Диагональные элеменнты равны $2 \dot N$, все остальные элементы матрицы равны 1. 

Вектор правых частей **b** определяется формулой $b_i=\frac{n \cdot (n+1)}{2} +i \cdot (2 \cdot (n-1))$, если элементы нумеруются с 1 и $b_i=\frac{n \cdot (n+1)}{2} +(i+1) \cdot (2 \cdot (n-1))$, если элементы нумеруются с 0. 

Матрица A обладает следующими особенностями:
1. Матрица A имеет диагональное преобладание, что позволяет использовать её в качестве тестовой для итерационных методов (Зейделя, Якоби и т. д.).
2. Заранее известно решение системы любой размерности x=(1,2,3,4,5...).

````python
import numpy
import sys
def printf(format, *args):
    sys.stdout.write(format % args)
N=int(input('N='));
A=numpy.ones([N,N]);
b=numpy.ones(N);
for i in range(0,N):
    for j in range(0,N):
        if (i==j):
            A[i,j]=2*N
        else:
            A[i,j]=1
    b[i]=N*(N+1)/2+(i+1)*(2*N-1)
x=numpy.linalg.solve(A,b)
printf("Корни уравнения\n")
for i in range(0,N):
    printf("%5.2f\t",x[i])
r=numpy.dot(A,x)-b
printf("\nМассив ошибок\n")
for i in range(0,N):
    printf("%.15e\t",r[i])
````

**Задача 3.** Написать программу найхождения определителя матрицы, вычисления обратной матрицы, нахождения собственные значений и собственные векторов матрицы.

````python
import numpy
import sys
def printf(format, *args):
    sys.stdout.write(format % args)
N=int(input('N='));
A=numpy.ones([N,N]);
for i in range(0,N):
    for j in range(0,N):
        if (i==j):
            A[i,j]=2*N
        else:
            A[i,j]=1
printf("Исходная матрица А\n");
for i in range(0,N):
    for j in range(0,N):
        printf("%8.2f\t",A[i][j])
    printf("\n");
a1=numpy.linalg.inv(A)
printf("Обратная матрица к матрице А=\n");
for i in range(0,N):
    for j in range(0,N):
        printf("%8.5f\t",a1[i][j])
    printf("\n");
printf('Проверка=\n');
a2=numpy.dot(A,a1)
for i in range(0,N):
    for j in range(0,N):
        printf("%8.5f\t",a2[i][j])
    printf("\n");
dt=numpy.linalg.det(A)
printf('Определитель=%8.4f\n',dt)
v=numpy.linalg.eig(A)
a3=v[0]
a4=v[1]
printf('Собственные числа=')
for j in range(0,N):
    printf("%8.5f\t",a3[j])
printf('\nСобственные вектора=\n');
for i in range(0,N):
    for j in range(0,N):
        printf("%8.5f\t",a4[i][j])
    printf("\n");    
````
